```python
from keras.preprocessing.image import img_to_array
from keras.models import load_model
model = load_model('model_file/all_model_class.h5')
# 클래스 리스트 선언
class_list = ['check','herringbone','solid','stripe', 'window']
import io
from PIL import Image
path = './design/1231231234.jpg'
import numpy as np
import cv2
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
import math
import matplotlib.pyplot as plot
from sklearn.cluster import MiniBatchKMeans

%matplotlib inline
import cv2
red_list = ['darkred',
'crimson',
'darksalmon',
'firebrick',
'indianred',
'lightcoral',
'maroon',
'salmon',
'tomato']
orange_list = ['coral',
'darkorange',
'lightsalmon',
'orange',
'orangered',
'papayawhip',
'peachpuff']
yellow_list = ['lemonchiffon',
'lightgoldenrodyellow',
'lightyellow',
'moccasin',
'navajowhite',
'palegoldenrod',
'wheat',
'yellow']
lg_list = ['aquamarine',
'chartreuse',
'darkkhaki',
'darkseagreen',
'greenyellow',
'lawngreen',
'lightgreen',
'lime',
'mediumspringgreen',
'olive',
'olivedrab',
'palegreen',
'springgreen',
'yellowgreen']
gre_list = ['darkgreen','darkolivegreen','forestgreen','green','honeydew','limegreen','mediumseagreen','seagreen']
chung_list = ['azure','cadetblue','darkcyan','darkslategrey','darkturquoise','lightcyan','lightseagreen','mediumaquamarine','mediumturquoise','teal','turquoise']
pa_list = ['aqua', 'blue','cornflowerblue','cyan','darkblue', 'deepskyblue','dodgerblue','lightblue','lightskyblue','lightsteelblue','mediumblue','paleturquoise','powderblue','darkslategray','royalblue','skyblue','steelblue']
nam_list = ['midnightblue','navy']
bo_list = ['thistle','slateblue','purple','plum','mediumslateblue','mediumpurple','mediumorchid','lavender','indigo','darkviolet','darkslateblue','darkorchid','darkmagenta','blueviolet']
ja_list = ['mediumvioletred']
pi_list = ['violet','pink','palevioletred','orchid','mistyrose','magenta','lightpink','lavenderblush','hotpink','fuchsia','deeppink']
br_list = ['brown','burlywood','chocolate','khaki','peru','rosybrown','saddlebrown','sandybrown','sienna','tan']
wh_list = ['floralwhite','ghostwhite','ivory','linen','mintcream','oldlace','seashell','snow','white','whitesmoke','cornsilk','blanchedalmond','bisque']
gr_list = ['darkgray','dimgray','gainsboro','gray','lightgray','lightslategray','silver','slategray']
bl_list = ['black']




with open(path, 'rb') as f:
    data = f.read()
img = Image.open(io.BytesIO(data)).convert("RGB")

img = img.resize((256, 256))
img = img_to_array(img)
img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))

pred = model.predict(img)
label = pred.argmax()
d = class_list[label]
from keras.preprocessing.image import img_to_array
from keras.models import load_model
model = load_model('model_file/model_class_design.h5')
# 클래스 리스트 선언
class_list = ['inner','outer','pants','top']
with open(path, 'rb') as f:
    data = f.read()
img = Image.open(io.BytesIO(data)).convert("RGB")

img = img.resize((256, 256))
img = img_to_array(img)
img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))

pred = model.predict(img)
label1 = pred.argmax()
img = cv2.imread(path)
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image = image.reshape((image.shape[0] * image.shape[1], 3)) #3차원이 2차원
k = 5 # 예제는 5개로 나누겠습니다
clt = KMeans(n_clusters = k)
clt.fit(image)

def centroid_histogram(clt):
    # grab the number of different clusters and create a histogram
    # based on the number of pixels assigned to each cluster
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)

    # normalize the histogram, such that it sums to one
    hist = hist.astype("float")
    hist /= hist.sum()

    # return the histogram
    return hist


hist = centroid_histogram(clt)
  


sum = 0
for i in range(4):
    try:
        if np.all(clt.cluster_centers_[i] > 250):
            hist = np.delete(hist,i)
            clt.cluster_centers_ = np.delete(clt.cluster_centers_,i,0)
    except:
        pass

for i in range(len(hist)):
    sum = sum + hist[i]  
for i in range(len(hist)):
    hist[i] = hist[i]/sum 

a = clt.cluster_centers_[np.argmax(hist)]
import webcolors

def closest_colour(requested_colour):
    min_colours = {}
    for key, name in webcolors.CSS3_HEX_TO_NAMES.items():
        r_c, g_c, b_c = webcolors.hex_to_rgb(key)
        rd = (r_c - requested_colour[0]) ** 2
        gd = (g_c - requested_colour[1]) ** 2
        bd = (b_c - requested_colour[2]) ** 2
        min_colours[(rd + gd + bd)] = name
    return min_colours[min(min_colours.keys())]

def get_colour_name(requested_colour):
    try:
        closest_name = actual_name = webcolors.rgb_to_name(requested_colour)
    except ValueError:
        closest_name = closest_colour(requested_colour)
        actual_name = None
    return actual_name, closest_name

requested_colour = (int(a[0]), int(a[1]), int(a[2]))
actual_name, closest_name = get_colour_name(requested_colour)
if closest_name in red_list:
    closest_name = '빨강'
elif closest_name in orange_list:
    closest_name = '주황'
elif closest_name in yellow_list:
    closest_name = '노랑'
elif closest_name in lg_list:
    closest_name = '연두'
elif closest_name in gre_list:
    closest_name = '초록'
elif closest_name in chung_list:
    closest_name = '청록'
elif closest_name in pa_list:
    closest_name = '파랑'
elif closest_name in nam_list:
    closest_name = '남색'
elif closest_name in bo_list:
    closest_name = '보라'
elif closest_name in ja_list:
    closest_name = '자주'
elif closest_name in pi_list:
    closest_name = '분홍'
elif closest_name in br_list:
    closest_name = '브라운'
elif closest_name in wh_list:
    closest_name = '흰색'
elif closest_name in gr_list:
    closest_name = '회색'
elif closest_name in bl_list:
    closest_name = '검정'






print (", closest colour name:", closest_name)
print('pattern :'+d+'     design :'+class_list[label1])
```

```python
def plot_colors(hist, centroids):

    # initialize the bar chart representing the relative frequency
    # of each of the colors
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0

    # loop over the percentage of each cluster and the color of
    # each cluster
    for (percent, color) in zip(hist, centroids):
        # plot the relative percentage of each cluster
        endX = startX + (percent * 300)
        cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX

    # return the bar chart
    return bar



bar = plot_colors(hist,clt.cluster_centers_)


# show our color bart
plt.figure()
plt.axis("off")


plt.imshow(bar)
```
